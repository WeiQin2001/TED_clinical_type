```python
import scanpy as sc
import scvelo as scv
import matplotlib.pyplot as plt
import omicverse as ov
import matplotlib
ov.plot_set()
```

```python
adata_CD8T = sc.read("../data/adata_CD8T_adata.h5ad")
adata_CD8T
```

```python
adata_CD8T.obs['clusters']=adata_CD8T.obs['CD8T_celltype']
adata_CD8T
```

```python
%matplotlib inline
ov.utils.embedding(adata_CD8T,basis='X_umap_CD8T',
                   color=['clusters','group'],wspace=0.8,
                   frameon='small',cmap='RdBu_r')
plt.savefig('./output/1.CD8T_umap.pdf', dpi=600)
plt.show()
```

**Slingshot**

```python
Traj=ov.single.TrajInfer(adata_CD8T,basis='X_umap_CD8T',groupby='clusters',
                         use_rep='X_scVI',n_comps=15)
Traj.set_origin_cells('CD8T_c04-LEF1-CCR7')
```

```python
matplotlib.use('Agg')
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 8))
Traj.inference(method='slingshot',num_epochs=1,debug_axes=axes)
plt.savefig('./output/slingshot_lineage_vessels.pdf', dpi=600)
plt.show()
```
