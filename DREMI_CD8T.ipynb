```python
import scanpy as sc
import omicverse as ov
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scprep

# Set up plotting parameters
ov.plot_set(font_path='Arial')
```

```python
adata_CD8T = sc.read("../data/adata_CD8T_adata.h5ad")
adata_CD8T
```

```python
adata_CD8T_magic = adata_CD8T.raw.to_adata().copy()
adata_CD8T_magic.X = adata_CD8T_magic.uns['magic_imputed_data_CD8T']
adata_CD8T_magic
```

```python
%matplotlib inline
ov.utils.embedding(adata_CD8T_magic,
                basis='X_umap_CD8T',frameon='small',wspace=0.5,
                color=['IKZF2'],show=False)
plt.savefig('./IKZF2/1.IKZF2_expression_magic.pdf', dpi=600)
plt.show()
```

```python
ov.pl.calculate_gene_density(adata_CD8T,
                              features=['IKZF2'],
                              basis='X_umap_CD8T',
                              dims=(0,1),
                              adjust=0.2,#adjust the bandwidth of the kernel density estimate
                              min_expr=0.01
                              )
ax=ov.pl.embedding(adata_CD8T,
                basis='X_umap_CD8T',
                color=['density_IKZF2'],#color_map="BuPu",
                frameon='small',
                vmax='p99.2',
                show=False,)
plt.savefig('./IKZF2/2.IKZF2_expression_density.pdf', dpi=600)
plt.show()
```

```python
import scprep
target_genes = ['KIR2DS4', 'KIR3DL2', 'NCR1','PIK3R1',"LYN",
                'PRF1', "GZMB","IL2RB",  "FCGR3A","NCAM1"]
IKZF2_expr = adata_CD8T_magic[:, 'IKZF2'].X.toarray().flatten()

for gene in target_genes:
    target_expr = adata_CD8T_magic[:, gene].X.toarray().flatten()
    dremi_score = scprep.stats.knnDREMI(
        IKZF2_expr, 
        target_expr,
        plot=True,
        filename=f'./IKZF2/{gene}.pdf'
    )
    print(f"IKZF2-{gene} DREMI score: {dremi_score}")
```
