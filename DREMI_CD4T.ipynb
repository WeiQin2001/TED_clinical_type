```python
import scanpy as sc
import omicverse as ov
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scprep

# Set up plotting parameters
ov.plot_set(font_path='Arial')
```

```python
adata_CD4T = sc.read("../data/adata_CD4T_adata.h5ad")
adata_CD4T
```

```python
adata_CD4T_magic = adata_CD4T.raw.to_adata().copy()
adata_CD4T_magic.X = adata_CD4T_magic.uns['magic_imputed_data_CD4T']
adata_CD4T_magic
```

```python
%matplotlib inline
ov.utils.embedding(adata_CD4T_magic,
                basis='X_diffmap',frameon='small',wspace=0.5,#color_map="BuPu",
                color=['TBX21'],show=False)
plt.savefig('./TBX21/1.TBX21_expression.pdf', dpi=600)
plt.show()
```

```python
ov.pl.calculate_gene_density(adata_CD4T,
                              features=['TBX21'],
                              basis='X_diffmap',
                              dims=(0,1),
                              adjust=0.2,#adjust the bandwidth of the kernel density estimate
                              min_expr=0.01
                              )
ax=ov.pl.embedding(adata_CD4T,
                basis='X_diffmap',
                color=['density_TBX21'],#color_map="BuPu",
                frameon='small',
                vmax='p99.2',
                show=False,)
plt.savefig('./TBX21/2.TBX21_expression_density.pdf', dpi=600)
plt.show()
```

```python
import scprep
target_genes = ["GZMA", 'GZMB', 'GZMH', "GZMM","IFNG",
                "TNF", "FASLG", 'GNLY', 'PRF1', 'NKG7']

tbx21_expr = adata_CD4T_magic[:, 'TBX21'].X.toarray().flatten()

for gene in target_genes:
    target_expr = adata_CD4T_magic[:, gene].X.toarray().flatten()
    dremi_score = scprep.stats.knnDREMI(
        tbx21_expr, 
        target_expr,
        plot=True,
        filename=f'./TBX21/{gene}.pdf'
    )
    print(f"TBX21-{gene} DREMI score: {dremi_score}")
```
